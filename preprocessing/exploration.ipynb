```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
ufo_500 = pd.read_json("ufo_500.json")
ufo_3000 = pd.read_json("ufo_3000.json")
ufo_all = pd.read_json("ufo_all.json")

ufo_500.describe()
```

```python
def emptyValues(df):
    print("UFO dataframe, nb records :",len(df))
    print("City, empty values",len(df[df["city"] == ""]))
    print("Comments, null values",len(df[df["comments"] == ""]))
    print("Country, empty values",len(df[df["country"] == ""]))
    print("Datetime, empty values",len(df[df["datetime"] == ""]))
    print("Duration (seconds), null values",len(df[df["duration (seconds)"] == 0]))
    print("Latitude, null values",len(df[df["latitude"] == 0.]))
    print("Longitude, null values",len(df[df["longitude"] == 0.]))
    print("Shape, null values",len(df[df["shape"] == ""]))
    print("State, empty values",len(df[df["state"] == ""]))
    print()
    
emptyValues(ufo_500)
emptyValues(ufo_3000)
emptyValues(ufo_all)
```

```python
def uniqueValues(df):
    print("UFO dataframe, nb records :",len(df))
    print("Country (nb " + str(len(df["country"].unique())) + ") :", df["country"].unique())
    print("Shape (nb " + str(len(df["shape"].unique())) + ") :", df["shape"].unique())
    print()
    
uniqueValues(ufo_500)
uniqueValues(ufo_3000)
uniqueValues(ufo_all)
```

```python
def outliers(df):
    print("UFO dataframe, nb records :",len(df))
    try:
        print("Duration (seconds), outliers :",len(df[df["duration (seconds)"] < 0]))
    except:
        print("Seems to have a prob with Duration type of data")
    try:
        print("Latitude, outliers :",len(df[abs(df["latitude"]) > 90]))
    except:
        print("Seems to have a prob with Latitude type of data")
    try:
        print("Longitude, outliers :",len(df[abs(df["longitude"]) > 180]))
    except:
        print("Seems to have a prob with Longitude type of data")
    print()

outliers(ufo_500)
outliers(ufo_3000)
outliers(ufo_all)
```

```python
converted_durations = pd.to_numeric(ufo_all["duration (seconds)"], errors="coerce")
print("Number of duration conversion errors :",converted_durations.isnull().sum())
print()

converted_latitudes = pd.to_numeric(ufo_all["latitude"], errors="coerce")
print("Number of latitude conversion errors :",converted_latitudes.isnull().sum())
```

```python
def datetimeErrors(df):
    converted_df = pd.to_datetime(df["datetime"], errors="coerce")
    
    print("UFO dataframe, nb records :",len(df))
    print("Number of datetime conversion errors :",converted_df.isnull().sum())
    print()
                             
datetimeErrors(ufo_500)
datetimeErrors(ufo_3000)
datetimeErrors(ufo_all)
```

```python
def dataGroupedByCountry(df):
    print("UFO dataframe, nb records :",len(df))
    print(df.groupby(['country']).count()['city'])
    print()

dataGroupedByCountry(ufo_500)
dataGroupedByCountry(ufo_3000)
dataGroupedByCountry(ufo_all)
```

```python
def nbEmptyStates(df, country):
    print("UFO dataframe","("+country+"),","nb records :",len(df[df["country"] == country]))
    print("nb of empty values for state :",len(df[(df["country"] == country) & (df["state"] == "")]))
    print()

nbEmptyStates(ufo_500, "us")
nbEmptyStates(ufo_3000, "us")
nbEmptyStates(ufo_all, "us")
```

```python
def nbEmptyStates(df, country):
    df_us = df[df["country"] == country]
    print("UFO dataframe","("+country+"),","nb records :",len(df[df["country"] == country]))
    print("list of states (" + str(len(df_us["state"].unique())) + ") :", sorted(df_us["state"].unique()))
    print()

nbEmptyStates(ufo_500, "us")
nbEmptyStates(ufo_3000, "us")
nbEmptyStates(ufo_all, "us")
```
